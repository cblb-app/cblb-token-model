```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 3, figsize=(15,3))

length = 100000 # checkin times

checkIn = np.arange(length)
tokenDistributed = 144600/(checkIn + 100) # Token distributed after each check in
tokenTotalSupply = np.arange(length, dtype=float) # Token total issued
tokenMaticRatio = np.arange(length, dtype=float) # Token total issued
fee = np.arange(length, dtype=float)

ax[0].set_title('Token Distributed Chart')
ax[0].set_xlabel('Check in times')
ax[0].set_ylabel('Token Distributed')

ax[1].set_title('Token Total Supply Chart')
ax[1].set_xlabel('Check in times')
ax[1].set_ylabel('Total supply')

ax[2].set_title('Expect CBLB/MATIC ratio Chart')
ax[2].set_xlabel('Check in times')
ax[2].set_ylabel('CBLB/MATIC ratio')

for i in checkIn:
    tokenTotalSupply[i] = np.sum(tokenDistributed[0:i])
    fee[i] = 0.11314554  # The fee for each check in is 144600 / (100 + 100) / 6390 = 0.11314554 MATIC
    if(i == 0):
        tokenMaticRatio[i] = 0
    else:
        tokenMaticRatio[i] = np.sum(fee[0:i])/np.sum(tokenTotalSupply[i])
        
print('Total checkIn: ', length)
print('Total supply: ', np.sum(tokenDistributed), 'CBLB')
print('Total fee acquire: ', np.sum(fee), 'MATIC') 
print('Average price CBLB/MATIC: ', np.sum(fee)/np.sum(tokenDistributed))

ax[0].plot(checkIn, tokenDistributed)
ax[1].plot(checkIn, tokenTotalSupply)
ax[2].plot(checkIn, tokenMaticRatio)
```
